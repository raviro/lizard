```python
from qiskit import *
%matplotlib inline
from qiskit import QuantumCircuit, QuantumRegister, ClassicalRegister, Aer, execute
from qiskit.tools.visualization import circuit_drawer
from qiskit.visualization import plot_histogram
simulator = Aer.get_backend('qasm_simulator')
import Our_Qiskit_Functions as oq
import numpy as np

def ksgen():
    for i in range(10):
        qc.cx(nfsr1[(0+i)%3],ks[i])
        qc.cx(nfsr1[(1+i)%3],ks[i])
        qc.ccx(nfsr2[(1+i)%5],nfsr2[(3+i)%5],ks[i])
        qc.ccx(nfsr1[(2+i)%3],nfsr2[(4+i)%5],ks[i])
        
        qc.cx(nfsr1[(0+i)%3],nfsr2[(0+i)%5])
        qc.ccx(nfsr2[(1+i)%5],nfsr2[(2+i)%5],nfsr2[(0+i)%5])
        qc.ccx(nfsr2[(3+i)%5],nfsr2[(4+i)%5],nfsr2[(0+i)%5])
        
        qc.ccx(nfsr1[(1+i)%3],nfsr1[(2+i)%3],nfsr1[(0+i)%3])
        
def ksgeninv():
    for i in reversed(range(10)):
        qc.ccx(nfsr1[(1+i)%3],nfsr1[(2+i)%3],nfsr1[(0+i)%3])
        
        qc.ccx(nfsr2[(3+i)%5],nfsr2[(4+i)%5],nfsr2[(0+i)%5])
        qc.ccx(nfsr2[(1+i)%5],nfsr2[(2+i)%5],nfsr2[(0+i)%5])
        qc.cx(nfsr1[(0+i)%3],nfsr2[(0+i)%5])
        
        qc.ccx(nfsr1[(2+i)%3],nfsr2[(4+i)%5],ks[i])
        qc.ccx(nfsr2[(1+i)%5],nfsr2[(3+i)%5],ks[i])
        qc.cx(nfsr1[(1+i)%3],ks[i])
        qc.cx(nfsr1[(0+i)%3],ks[i])
    
    
    
    
nfsr1 = QuantumRegister(3)
nfsr2 = QuantumRegister(5)
ks = QuantumRegister(10)
trg = QuantumRegister(1)
cl = ClassicalRegister(8)

qc = QuantumCircuit(nfsr1,nfsr2,ks,trg,cl)

for i in range(3):
    qc.h(nfsr1[i])
    
for i in range(5):
    qc.h(nfsr2[i])
    
qc.x(trg[0])



for i in range(16):
    ksgen()
    Grover_Oracle([0,1,0,1,0,0,1,1,1,0], qc, [ks[0],ks[1],ks[2],ks[3],ks[4],ks[5],ks[6],ks[7],ks[8],ks[9]], trg)
    
    ksgeninv()
    Grover_Diffusion([0,0,0,0,0,0,0,0],qc, [nfsr1[0],nfsr1[1],nfsr1[2],
                                            nfsr2[0],nfsr2[1],nfsr2[2],nfsr2[3],nfsr2[4]], trg)
    
for i in range(3):
    qc.measure(nfsr1[i],cl[i])

for i in range(5):
    qc.measure(nfsr2[i],cl[i+3])

    
job = execute(qc, simulator, shots=1024)
result = job.result()
counts = result.get_counts(qc)
print(counts)
plot_histogram(counts) 
```
