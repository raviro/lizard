```python
from qiskit import *
%matplotlib inline
from qiskit import QuantumCircuit, QuantumRegister, ClassicalRegister, Aer, execute
from qiskit.tools.visualization import circuit_drawer
from qiskit.visualization import plot_histogram
simulator = Aer.get_backend('qasm_simulator')
import Our_Qiskit_Functions as oq
import numpy as np


def toffoli7(a,b,c,d,e,f,g,h,trg):
    qc.ccx(a,b,anc[0])
    qc.ccx(c,anc[0],anc[1])
    qc.ccx(d,anc[1],anc[2])
    qc.ccx(e,anc[2],anc[3])
    qc.ccx(f,anc[3],anc[4])
    qc.ccx(g,anc[4],anc[5])
    qc.ccx(h,anc[5],anc[6])
    qc.cx(anc[5],trg)
    qc.ccx(h,anc[5],anc[6])
    qc.ccx(g,anc[4],anc[5])
    qc.ccx(f,anc[3],anc[4])
    qc.ccx(e,anc[2],anc[3])
    qc.ccx(d,anc[1],anc[2])
    qc.ccx(c,anc[0],anc[1])
    qc.ccx(a,b,anc[0])
    
    
def func(arr,trg):
    toffoli7(arr[0],arr[1],arr[2],arr[3],arr[4],arr[5],arr[6],arr[7],trg[0])
    for i in range(8):
        qc.cx(trg[0],arr[i])
        
def funcinv(arr,trg):
    for i in range(8):
        qc.cx(trg[0],arr[i])
    toffoli7(arr[0],arr[1],arr[2],arr[3],arr[4],arr[5],arr[6],arr[7],trg[0])
    
arr = QuantumRegister(8)
anc = QuantumRegister(7)
trg = QuantumRegister(1)

target = QuantumRegister(1)

anc1 = QuantumRegister(7)
cl = ClassicalRegister(8)

qc = QuantumCircuit(arr,anc,trg,target,anc1,cl)

qc.h(arr)
qc.x(target[0])

for i in range(16):
    func(arr,trg)
    oq.Grover_Oracle([0,0,0,0,0,0,0,0,1], qc, [arr[0],arr[1],arr[2],arr[3],arr[4],arr[5],arr[6],arr[7],trg[0]], target, anc1)
    funcinv(arr,trg)
    oq.Grover_Diffusion([0,0,0,0,0,0,0,0], qc, arr, target, anc1)
    
for i in range(8):
    qc.measure(arr[i],cl[i])
        
job = execute(qc, simulator, shots=1024)
result = job.result()
counts = result.get_counts(qc)
print(counts)
plot_histogram(counts) 
```
